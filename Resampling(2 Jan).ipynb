```python
from sklearn.datasets import make_classification
from sklearn.utils import resample
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_classification(n_samples=1000, n_features=10,n_classes=2, weights=[0.8, 0.2])
```

```python
data
```

```python
X = data[0]
y = data[1]
```

```python
df = pd.DataFrame(X, columns=["col_{}".format(i) for i in range(len(X[0]))])
```

```python
df["target"] = y
```

```python
df.head()
```

```python
# Down-Sampling --> Deleting rows from big class
# Data become less

# Up-Sampling --> Duplicating rows from small class
# Data become double
```

```python
zero = df[df["target"]==0]
one = df[df["target"]==1]
```

```python
down = resample(zero, replace=False, n_samples=len(one))
```

```python
down
```

```python
down_data = pd.concat([one, down], axis=0)
```

```python
down_data
```

```python
up = resample(one, replace=True, n_samples=len(zero))
```

```python
up
```

```python
up_data = pd.concat([zero, up], axis=0)
```
